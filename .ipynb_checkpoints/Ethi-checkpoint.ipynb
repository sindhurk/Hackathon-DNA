```python
driver = get_driver()
weblinks = []
count = 0
company_details_backup = company_details
companies = {'SilkAir': {'website': 'www.silkair.com'}}
```

```python
for key, value in company_details_backup.items():
    try:
        if count == 10:
            break
        count+=1
        website_url = value["website"].split()[0]
        if website_url != "NA":
            soup = get_soup_data(driver, website_url)
            anchors = soup.select("a[href]")
            for anchor_tag in anchors:
                if anchor_tag["href"] != "#" and anchor_tag.text != "":
                    weblinks.append(anchor_tag["href"])
    except:
        print(key)
        print("Error processing")
        continue
driver.quit()
unique_links = list(set(weblinks))
for link in unique_links:
    if "product" in link or "service" in link or "solution" in link:
        print(link)
```
