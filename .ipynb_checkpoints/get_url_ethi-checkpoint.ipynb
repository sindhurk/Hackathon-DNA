```python
import pandas as pd
from selenium import webdriver
import time
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from selenium.webdriver.firefox.options import Options
from collections import *

content = pd.read_excel("/home/ethi/Downloads/hackathon.xlsx",sheetname="Training Data")
company_names = content["Company Name"]

def get_driver():
    options = Options()
    options.add_argument("--headless")
    return webdriver.Firefox(executable_path="/home/ethi/Downloads/geckodriver",firefox_options=options)
    #Chrome
#     options = webdriver.ChromeOptions()
#     options.add_argument(" -- incognito")
    
#     options.add_argument('--ignore-certificate-errors')
#     options.add_argument("--test-type")
#     options.add_argument("--headless")
#     return webdriver.Chrome(executable_path="/usr/lib/chromium-browser/chromedriver", chrome_options=options)

def get_soup_data(url):
    driver.get(url)
    html = driver.page_source
    return BeautifulSoup(html, 'html.parser')

def get_company_url_from_wiki(url):
    soup = get_soup_data(url)
    infobox = soup.find("table",{"class":"infobox"})
    if infobox:
        for tr in infobox.find_all("tr")[1:]:
            headers = tr.find_all("th")
            values = tr.find_all("td")
            if len(headers) == 1 and len(values) == 1 and headers[0].text == "Website": 
                return values[0].text
    return url        
driver = get_driver()
count = 1
google_url = "https://www.google.com/search?q="
duck_duck_go_url = "https://duckduckgo.com/?t=hf&ia=web&q="

def get_company_url(search_engine, url):
    soup = get_soup_data(url)
    if search_engine == google_url:
        return soup.find("div",{"id":"search"}).find_all("cite")[0].text
    # DuckDuckGo
    first_link = soup.find("div",{"id":"r1-0"})
    company_domain = first_link["data-domain"]
    company_url = soup.find("span",{"class":"result__url__domain"}).text
    return company_url

def clean_url(url):
    url = url.replace("https://", "").replace("www.", "")
    if url[-1] == "/":
        url = url[:-1]
    return url

def search_engine_search(search_engine, company):
    complete_url_website = search_engine + company + " website"
    complete_url = search_engine + company
    url1 = clean_url(get_company_url(search_engine, complete_url))
    url2 = clean_url(get_company_url(search_engine, complete_url_website))
    return url1, url2

company_websites = {}
for company in company_names:
    url1, url2 = search_engine_search(duck_duck_go_url,company)
    if url1 == url2:
        company_websites[company] = url1
        if url1 == "en.wikipedia.org":
            get_company_url_from_wiki()
            
    else:
        url3, url4 = search_engine_search(google_url,company)
        if url3 == url4:
            company_websites[company] = url3
        else:
            websites=[url1, url2, url3, url4]
            fre_web = Counter(websites).most_common(1)[0]
            if fre_web[1] > 1:
                company_websites[company] = fre_web[0]
            else:
                company_websites[company] = websites
driver.quit()
        
```
